```python
from ultralytics import YOLO
import wandb

wandb.init(project="amorphous-yolo-phase0", name="coco8_baseline_colab")

model = YOLO("yolo26n.pt")  # or yolov8n.pt if your ultralytics is v8

results = model.train(
    data="coco8.yaml",   # built-in tiny dataset
    epochs=3,
    imgsz=640,
    project="experiments",
    name="phase0_coco8_colab",
    device=0,            # use Colab GPU
)

wandb.finish()
print("Training done:", results)
```
